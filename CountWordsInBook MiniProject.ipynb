### START COUNT THE WORDS IN THE BOOK PROJECT

```python
input = sc.textFile("file:///home/bigdata/workspace/Spark_resources/Book")
```

```python
bookWords = input.flatMap(lambda x: x.split())
```

```python
bookWordsCollect = bookWords.collect()
```

```python
bookWordsCount = bookWords.countByValue()
```

```python
print(bookWordsCount)
```

```python
for word,count in bookWordsCount.items():
    cleanWord  = word.encode('ascii','ignore')
    print(cleanWord)
```

* here encode method is used to clean all the errors i.e utf-8 or unicode and convert them into ASCII and ignore the errors, which occurs during the conversion process.

```python
for word,count in bookWordsCount.items():
    cleanWord  = word.encode('ascii','ignore')
    if (cleanWord):
        print(cleanWord.decode() + " " + str(count))
```

## Improving the word count script with regular expressions.

```python
# import packages for regular expression
import re
```

```python
def normalizeWords(text):
    return re.compile(r'\W+', re.UNICODE).split(text.lower())
```

* re.compile() is used to remove the punctutation, commas and captitalization from the book.

```python
input = sc.textFile("file:///home/bigdata/workspace/Spark_resources/Book")
```

```python
words = input.flatMap(normalizeWords)
```

```python
wordsCollect = words.collect()
```

```python
print(wordsCollect)
```

```python
wordCounts = words.countByValue()
```

```python
for word, count in wordCounts.items():
    cleanWord = word.encode('ascii', 'ignore')
    if (cleanWord):
        print(cleanWord.decode() + " " + str(count))
```

## SORTING THE RESULTS: step 1

* we could sort what countBYValue() returns, but let's use RDD's to keep it scalable.
* First, do countByValue() the hard way.

```python
wordCounts = words.map(lambda x: (x,1))
```

* Convert each word to a key/value pair with a value of 1

```python
wordCountsCollect = wordCounts.collect()
```

```python
print(wordCountsCollect)
```

```python
wordCountsReduce = wordCounts.reduceByKey(lambda x,y: x + y)
```

* THen count them all up with reduceByKey()

```python
wordCountsReduceCollect = wordCountsReduce.collect()
```

```python
print(wordCountsReduceCollect)
```

### SORTING THE RESULTS: STEP 2


* Flip the (word,count) pairs to (count, word)
* Then use sortByKey() to sort by count (now that count is our new key)

```python
wordCountsSorted = wordCountsReduce.map(lambda x: (x[1], x[0]))
```

```python
wordCountsSortedCollect = wordCountsSorted.collect()
```

```python
print(wordCountsSortedCollect)
```

```python
wordCountsSortedSort = wordCountsSorted.sortByKey()
```

```python
results = wordCountsSortedSort.collect()
```

```python
for result in results:
    count = str(result[0])
    word = result[1].encode('ascii', 'ignore')
    if (word):
        print(word.decode() + ":\t\t" + count)
```
