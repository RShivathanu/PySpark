```python
import pyspark
```

```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setMaster("local").setAppName("CustomerSpendings")
```

```python
sc = SparkContext(conf=conf)
```

```python
lines = sc.textFile("file:///home/bigdata/workspace/Spark_resources/customer-orders.csv")
```

```python
def parseLine(line):
    fields = line.split(',')
    customerId = fields[0]
    dollarAmount = fields[2]
    return (int(customerId), float(dollarAmount))
```

```python
amount = lines.map(parseLine)
```

```python
amountCollect = amount.collect()
```

```python
print(amountCollect)
```

```python
totalByCustomer = amount.reduceByKey(lambda x,y: x + y)
```

```python
results = totalByCustomer.collect()
```

```python
for result in results:
    print(result)
```

```python
for result in results:
    print(result[0] , "\t{:.2f}".format(result[1]))
```

```python
flipped = totalByCustomer.map(lambda x : (x[1],x[0]))
```

```python
totalByCustomerSorted = totalByCustomer.sortByKey()
```

```python
results = totalByCustomerSorted.collect()
```

```python
for result in results:
    print(result[0],"\t{:.2f}".format(result[1]))
```
